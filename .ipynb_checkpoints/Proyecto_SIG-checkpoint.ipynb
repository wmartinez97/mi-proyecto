```python
print(src)  # Información del raster
print("Número de bandas:", src.count)  # Número de bandas
print("Tamaño (ancho x alto):", src.width, "x", src.height)  # Dimensiones
print("Tipo de dato:", src.dtypes[0])  # Tipo de dato
```

```python
    # Cambia los índices (1, 2, 3) si los canales RGB están en otras bandas
    red = src.read(1)  # Banda roja
    green = src.read(2)  # Banda verde
    blue = src.read(3)  # Banda azul
    
    # Función para manejar valores no válidos (NaN o infinitos)
    def clean_array(array):
        array = np.nan_to_num(array, nan=0.0, posinf=0.0, neginf=0.0)  # Reemplaza NaN e infinitos por 0
        return array

    # Normalizar los valores a un rango de 0-1 para visualización
    def normalize(array):
        array_min, array_max = array.min(), array.max()
        if array_max - array_min > 0:
            return (array - array_min) / (array_max - array_min)
        else:
            return array  # Evitar división por cero si todos los valores son iguales

    # Limpiar y normalizar las bandas
    red_clean = clean_array(red)
    green_clean = clean_array(green)
    blue_clean = clean_array(blue)

    red_norm = normalize(red_clean)
    green_norm = normalize(green_clean)
    blue_norm = normalize(blue_clean)
    
    # Combinar en una imagen RGB
    rgb_image = np.stack((red_norm, green_norm, blue_norm), axis=-1)

    # Visualizar la imagen RGB
    plt.figure(figsize=(10, 10))
    plt.imshow(rgb_image)
    plt.title("Visualización de la imagen RGB")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.axis('off')
    plt.show()
```

```python
    # Cambia los índices según el orden de las bandas en tu archivo
    nir = src.read(4)   # Banda de infrarrojo cercano (NIR)
    red = src.read(1)   # Banda roja
    green = src.read(2) # Banda verde

    # Función para manejar valores no válidos (NaN o infinitos)
    def clean_array(array):
        array = np.nan_to_num(array, nan=0.0, posinf=0.0, neginf=0.0)  # Reemplaza NaN e infinitos por 0
        return array

    # Normalizar los valores a un rango de 0-1 para visualización
    def normalize(array):
        array_min, array_max = array.min(), array.max()
        if array_max - array_min > 0:
            return (array - array_min) / (array_max - array_min)
        else:
            return array  # Evitar división por cero si todos los valores son iguales

    # Limpiar y normalizar las bandas
    nir_clean = clean_array(nir)
    red_clean = clean_array(red)
    green_clean = clean_array(green)

    nir_norm = normalize(nir_clean)
    red_norm = normalize(red_clean)
    green_norm = normalize(green_clean)
    
    # Combinar en una imagen de falso color (NIR, Red, Green)
    false_color_image = np.stack((nir_norm, red_norm, green_norm), axis=-1)

    # Visualizar la imagen de falso color
    plt.figure(figsize=(10, 10))
    plt.imshow(false_color_image)
    plt.title("Visualización de la imagen en Falso Color (NIR, Red, Green)")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.axis('off')
    plt.show()
```

```python

    # Cambia los índices según el orden de las bandas en tu archivo
    nir = src.read(4)   # Banda de infrarrojo cercano (NIR)
    red = src.read(1)   # Banda roja

    # Función para manejar valores no válidos (NaN o infinitos)
    def clean_array(array):
        array = np.nan_to_num(array, nan=0.0, posinf=0.0, neginf=0.0)  # Reemplaza NaN e infinitos por 0
        return array

    # Limpiar las bandas
    nir_clean = clean_array(nir)
    red_clean = clean_array(red)

    # Calcular el NDVI
    ndvi = (nir_clean - red_clean) / (nir_clean + red_clean)
    ndvi = np.nan_to_num(ndvi, nan=0.0, posinf=0.0, neginf=0.0)  # Manejo de posibles divisiones por 0

    # Visualizar el NDVI
    plt.figure(figsize=(10, 10))
    ndvi_cmap = plt.cm.viridis  # Colormap para el NDVI
    ndvi_cmap.set_bad(color='white')  # Color para valores no válidos
    plt.imshow(ndvi, cmap='RdYlGn', vmin=-1, vmax=1)  # NDVI tiene un rango de -1 a 1
    plt.colorbar(label="NDVI")
    plt.title("Mapa de NDVI")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.axis('off')
    plt.show()
```
